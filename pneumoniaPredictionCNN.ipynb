```python
from tensorflow.keras.preprocessing.image import ImageDataGenerator
import tensorflow as tf
```

```python
tf.__version__
```

```python
train_datagen = ImageDataGenerator(rescale = 1./255,
                                   shear_range = 0.2,
                                   zoom_range = 0.2,
                                   horizontal_flip = True)
training_set = train_datagen.flow_from_directory('/kaggle/input/chest-xray-pneumonia/chest_xray/train',
                                                 target_size = (64, 64),
                                                 batch_size = 32,
                                                 class_mode = 'binary')
```

```python
test_datagen = ImageDataGenerator(rescale = 1./255)
test_set = test_datagen.flow_from_directory('/kaggle/input/chest-xray-pneumonia/chest_xray/test',
                                            target_size = (64, 64),
                                            batch_size = 32,
                                            class_mode = 'binary')
```

```python
cnn = tf.keras.models.Sequential()
```

```python
cnn.add(tf.keras.layers.Conv2D(filters=32, kernel_size=3, activation='relu', input_shape=[64, 64, 3]))
```

```python
cnn.add(tf.keras.layers.MaxPool2D(pool_size=2, strides=2))
```

```python
cnn.add(tf.keras.layers.Conv2D(filters=32, kernel_size=3, activation='relu'))
cnn.add(tf.keras.layers.MaxPool2D(pool_size=2, strides=2))
```

```python
cnn.add(tf.keras.layers.Flatten())
```

```python
cnn.add(tf.keras.layers.Dense(units=128, activation='relu'))
```

```python
cnn.add(tf.keras.layers.Dense(units=1, activation='sigmoid'))
```

```python
cnn.compile(optimizer = 'adam', loss = 'binary_crossentropy', metrics = ['accuracy'])
```

```python
cnn.fit(x = training_set, validation_data = test_set, epochs = 25)
```

```python
import numpy as np
from keras.preprocessing import image
test_image = image.load_img('/kaggle/input/chest-xray-pneumonia/chest_xray/val/NORMAL/NORMAL2-IM-1442-0001.jpeg', target_size = (64, 64))
test_image = image.img_to_array(test_image)
test_image = np.expand_dims(test_image, axis = 0)
result = cnn.predict(test_image)
training_set.class_indices

```
